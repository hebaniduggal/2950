```python
#Python3
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO

ps2texts=urlopen('https://courses.cit.cornell.edu/info2950_2017sp/resources/ps2data.zip')
with ZipFile(BytesIO(ps2texts.read())) as ziptexts:
    ptexts=[ziptexts.read('ps2data/p{}.txt'.format(i)).decode('utf-8') for i in range(30)]
    stexts=[ziptexts.read('ps2data/s{}.txt'.format(i)).decode('utf-8') for i in range(30)]
    ttexts=[ziptexts.read('ps2data/t{}.txt'.format(i)).decode('utf-8') for i in range(10)]
ps2texts.close()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
import re
from scipy.stats import linregress
```

```python
alltexts = ptexts + stexts + ttexts
a = Counter()
distinct = 0
total = 0
distinct_words = []
total_count = []
distinct_count = []

for i in alltexts: 
    a = Counter(re.findall("[a-z']+",i.lower()))
    for b in a:
        if b in distinct_words:
            distinct_words = distinct_words
        else:
            distinct_words.append(b)
            distinct+=1
    for c in a.values():
        total += c
    distinct_count.append(distinct)
    total_count.append(total)
print('')
print(distinct_count)
print('')
print(total_count)

plt.figure(figsize=(10.5,5))
plt.subplot(1,2,1)
plt.xlabel('Total Words'), plt.ylabel('Distinct Words')
plt.plot(total_count,distinct_count,'.-')
logxdata,logydata = map(np.log,(total_count,distinct_count))
slope, intercept, r_value, p_value, std_err = linregress(logxdata,logydata)
plt.plot(total_count,np.exp(intercept)*total_count**slope,'r',lw=2)
plt.legend(['Corpus','Best Fit'],loc='upper left')

plt.subplot(1,2,2)
plt.xlabel('Total Words'), plt.ylabel('Distinct Words')
plt.loglog(total_count,distinct_count,'.-')
plt.xlim(500,150000)
plt.ylim(300,15000)
logxdata,logydata = map(np.log,(total_count,distinct_count))
slope, intercept, r_value, p_value, std_err = linregress(logxdata,logydata)
plt.plot(total_count,np.exp(intercept)*total_count**slope,'r',lw=2)
plt.legend(['Corpus','Best Fit, Slope = 0.6897'],loc='upper left')

plt.show()
```
